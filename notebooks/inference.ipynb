# Inference Notebook

Use this notebook to run the trained spaCy model on your own inputs without typing Python in the terminal.

Before using it, make sure you have already trained the model once so that `custom_query_derivation/` exists.

```python
from pathlib import Path

import spacy


def find_repo_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "main.py").exists():
            return path
    raise FileNotFoundError("Could not find the spaCy-NER repo root.")


repo_root = find_repo_root(Path.cwd())
model_path = repo_root / "custom_query_derivation"

if not model_path.exists():
    raise FileNotFoundError(
        "Trained model not found. Run `python3 main.py` first to create custom_query_derivation/."
    )

nlp = spacy.load(model_path)
print(f"Loaded model from {model_path}")
```

```python
def predict(text: str):
    doc = nlp(text)
    return {
        "text": text,
        "entities": [
            {
                "text": ent.text,
                "label": ent.label_,
                "start": ent.start_char,
                "end": ent.end_char,
            }
            for ent in doc.ents
        ],
    }
```

## Single input

Edit `text` and run the cell.

```python
text = "painting by John Idiot Smith"
predict(text)
```

## Multiple inputs

Edit the list below and run the cell.

```python
texts = [
    "Rembrandt drawing of a landscape",
    "Gucci leather bag",
    "Ancient Roman marble bust",
]

[predict(text) for text in texts]
```
