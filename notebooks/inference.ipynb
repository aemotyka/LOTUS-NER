# Inference Notebook

Use this notebook to run the trained spaCy model on your own inputs without typing Python in the terminal.

Before using it, make sure you have already trained the model once so that `models/consolidated_query_derivation/` exists.

```python
from pathlib import Path

import spacy


def find_repo_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "main.py").exists():
            return path
    raise FileNotFoundError("Could not find the spaCy-NER repo root.")


repo_root = find_repo_root(Path.cwd())
model_path = repo_root / "models" / "consolidated_query_derivation"

if not model_path.exists():
    raise FileNotFoundError(
        "Trained model not found. Run `python3 main.py` first to create models/consolidated_query_derivation/."
    )

nlp = spacy.load(model_path)
print(f"Loaded model from {model_path}")
```

```python
def predict(text: str):
    doc = nlp(text)
    return {
        "text": text,
        "entities": [
            {
                "text": ent.text,
                "label": ent.label_,
                "start": ent.start_char,
                "end": ent.end_char,
            }
            for ent in doc.ents
        ],
    }
```

## Single input

Edit `text` and run the cell.

```python
text = "An American Aesthetic Movement side table designed by Louis Comfort Tiffany and produced by Tiffany Studios, made in favrile glass, bronze, and inlaid wood. The side table has a lower shelf, slender cabriole supports, and a glowing inset top panel as major feature components. The decorative program depicts a luminous wisteria garden at twilight, which serves as the principal subject theme. This work belongs to the late Aesthetic Movement and reflects the broader influence of Japonisme in Gilded Age America, providing useful historical context. Its origin is American, its period is Aesthetic Movement, and its material/technique combines favrile glass, bronze, and inlaid wood. The named artist/designer is Louis Comfort Tiffany, and the associated brand is Tiffany Studios."

predict(text)
```

## Multiple inputs

Edit the list below and run the cell.

```python
texts = [
    "Chinese ivory framed reverse glass painting",
    "The Twelve Princesses by Gustave-Max Stevens",
    "Cleopatra by Julius Kronberg",
    "Oscar Heyman Colombian Emerald Ring, 2.60 Carats",
    "Tiffany & Co. Japanese Pattern Fruit Service",
]

[predict(text) for text in texts]
```
